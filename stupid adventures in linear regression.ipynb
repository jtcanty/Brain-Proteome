```r
returnObservations <- function(row) {
    
    observation1 <- c(row[[2]], 1, 0)
    observation2 <- c(row[[3]], 1, 0)
    observation3 <- c(row[[4]], 0, 0)
    observation4 <- c(row[[5]], 0, 0)
    observation5 <- c(row[[6]], 0, 1)
    observation6 <- c(row[[7]], 0, 1)
    observation7 <- c(row[[8]], 1, 1)
    observation8 <- c(row[[9]], 1, 1)
    
    observations <- list(observation1, observation2, observation3, observation4, observation5, observation6, observation7, observation8)
    return(observations)
}
```

```r
(acg1 <- returnObservations(acg1))
(acg2 <- returnObservations(acg2))
(acg3 <- returnObservations(acg3))
(acg4 <- returnObservations(acg4))
(acg5 <- returnObservations(acg5))
(fc1  <- returnObservations(fc1))
(fc2  <- returnObservations(fc2))
(fc3  <- returnObservations(fc3))
(fc4  <- returnObservations(fc4))
(fc5  <- returnObservations(fc5))
```

```r
reformatObservations <- function(batch) {
    
    responseVector <- c(batch[[1]][1], batch[[2]][1], batch[[3]][1], batch[[4]][1],
                        batch[[5]][1], batch[[6]][1], batch[[7]][1], batch[[8]][1])
    covariate1Vector <- c(batch[[1]][2], batch[[2]][2], batch[[3]][2], batch[[4]][2],
                          batch[[5]][2], batch[[6]][2], batch[[7]][2], batch[[8]][2])
    covariate2Vector <- c(batch[[1]][3], batch[[2]][3], batch[[3]][3], batch[[4]][3],
                          batch[[5]][3], batch[[6]][3], batch[[7]][3], batch[[8]][3])
    
    return(list(responseVector, covariate1Vector, covariate2Vector))
}
```
